```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import os

# =========================
# Save path
# =========================
save_dir = "./save_figs"
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, "active_H_exclusion.svg")

# =========================
# Figure settings
# =========================
plt.rcParams["font.family"] = "Arial"
plt.rcParams["svg.fonttype"] = "none"

fig, ax = plt.subplots(figsize=(3.2, 2.4), dpi=300)
ax.set_xlim(0, 10)
ax.set_ylim(0, 7)
ax.axis("off")

# =========================
# Colors
# =========================
bond_color = "#2F2F2F"
h_color = "#E85C4A"
cross_color = "#E85C4A"
text_color = "#222222"

# =========================
# Molecule 1: alcohol R-O-H
# =========================
ax.text(1.0, 5.2, "R", fontsize=16, ha="center", va="center", color=text_color)
ax.plot([1.4, 2.4], [5.2, 5.2], color=bond_color, lw=2)
ax.text(2.8, 5.2, "O", fontsize=16, ha="center", va="center", color=text_color)
ax.plot([3.2, 4.2], [5.2, 5.2], color=bond_color, lw=2)

h_circle = Circle((4.7, 5.2), 0.45, facecolor="#FBE3DF", edgecolor=h_color, lw=2)
ax.add_patch(h_circle)
ax.text(4.7, 5.2, "H", fontsize=15, fontweight="bold", ha="center", va="center", color=h_color)

# Red cross
ax.plot([5.7, 6.7], [4.7, 5.7], color=cross_color, lw=3)
ax.plot([5.7, 6.7], [5.7, 4.7], color=cross_color, lw=3)

# =========================
# Molecule 2: amide N-H
# =========================
ax.text(1.0, 2.2, "R", fontsize=16, ha="center", va="center", color=text_color)
ax.plot([1.4, 2.2], [2.2, 2.2], color=bond_color, lw=2)
ax.text(2.6, 2.2, "C", fontsize=16, ha="center", va="center", color=text_color)

# Carbonyl
ax.plot([2.8, 3.4], [2.5, 3.2], color=bond_color, lw=2)
ax.plot([3.0, 3.6], [2.4, 3.1], color=bond_color, lw=1.5)
ax.text(3.9, 3.3, "O", fontsize=16, ha="center", va="center", color=text_color)

# C-N-H
ax.plot([3.0, 4.0], [2.2, 2.2], color=bond_color, lw=2)
ax.text(4.4, 2.2, "N", fontsize=16, ha="center", va="center", color=text_color)
ax.plot([4.8, 5.6], [2.2, 2.2], color=bond_color, lw=2)

h_circle2 = Circle((6.1, 2.2), 0.45, facecolor="#FBE3DF", edgecolor=h_color, lw=2)
ax.add_patch(h_circle2)
ax.text(6.1, 2.2, "H", fontsize=15, fontweight="bold", ha="center", va="center", color=h_color)

# Red cross
ax.plot([7.0, 8.0], [1.7, 2.7], color=cross_color, lw=3)
ax.plot([7.0, 8.0], [2.7, 1.7], color=cross_color, lw=3)

# =========================
# Title / label
# =========================
ax.text(
    5.0, 6.45,
    "Active H exclusion",
    fontsize=15,
    fontweight="bold",
    ha="center",
    va="center",
    color=h_color
)

ax.text(
    5.0, 0.55,
    "Remove labile protic hydrogen groups",
    fontsize=10.5,
    ha="center",
    va="center",
    color=text_color
)

# =========================
# Save
# =========================
plt.savefig(save_path, format="svg", bbox_inches="tight", transparent=True)
plt.close()

print(f"Saved to: {save_path}")
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import os
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import rdMolDraw2D


def smiles_to_svg(
    smiles,
    save_path="./molecule.svg",
    width=320,
    height=240,
    kekulize=True
):
    """
    输入一个 SMILES，生成分子结构 SVG 文件
    """

    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        raise ValueError(f"Invalid SMILES: {smiles}")

    mol = Chem.Mol(mol)

    if kekulize:
        try:
            Chem.Kekulize(mol)
        except Exception:
            pass

    try:
        AllChem.Compute2DCoords(mol)
    except Exception:
        pass

    drawer = rdMolDraw2D.MolDraw2DSVG(width, height)
    opts = drawer.drawOptions()

    opts.clearBackground = False
    opts.bondLineWidth = 2.0
    opts.minFontSize = 12
    opts.maxFontSize = 18
    opts.padding = 0.08

    drawer.DrawMolecule(mol)
    drawer.FinishDrawing()

    svg = drawer.GetDrawingText()

    os.makedirs(os.path.dirname(save_path) or ".", exist_ok=True)

    with open(save_path, "w", encoding="utf-8") as f:
        f.write(svg)

    print(f"Saved SVG to: {save_path}")


if __name__ == "__main__":

    smiles = 'B(C1=C(C=CC2=CC=CC=C12)OC3CCOCC3)(O)O'

    save_name = f"{smiles}.svg"
    smiles_to_svg(
        smiles=smiles,
        save_path=save_name,
        width=320,
        height=240
    )
```
