```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# =========================
# 输入数据
# =========================
counts = [84, 55, 108, 131, 49, 78, 93]
count_ratio = [i / sum(counts) for i in counts]
print(count_ratio)

colors = [
    "#6C2C0A",
    "#C15416",
    "#D57846",
    "#F29D72",
    "#BE739D",
    "#6A2F77",
    "#4C6BA2"
]

# =========================
# 画 donut
# =========================
fig, ax = plt.subplots(figsize=(7, 7))

# 设置透明背景
fig.patch.set_alpha(0)
ax.set_facecolor("none")

explode = [0.03] * len(counts)

wedges, texts = ax.pie(
    counts,
    colors=colors,
    startangle=90,
    explode=explode,
    counterclock=False,
    wedgeprops=dict(
        width=0.25,
        edgecolor="none"   # 不要白色边框
    )
)

ax.set(aspect="equal")
ax.axis("off")  # 去掉坐标轴

plt.tight_layout(pad=0)

plt.savefig(
    "../plots/fig3_c.svg",
    dpi=600,
    bbox_inches="tight",
    pad_inches=0,
    transparent=True
)

plt.show()

fig3_c_df = pd.DataFrame({
    "prototype": [f"P{i+1}" for i in range(len(counts))],
    "count": counts,
    "ratio": count_ratio,
    "color": colors,
})

fig3_c_df.to_csv(
    "../plots/fig3_c.csv",
    index=False,
    encoding="utf-8-sig"
)
```
