```python
import matplotlib.pyplot as plt
import numpy as np

# =========================
# 输入数据
# =========================
# counts = [78, 131, 49, 55, 108, 84, 93]
counts = [84, 55, 108, 131, 49, 78, 93]
count_ratio = [i / sum(counts) for i in counts]
print(count_ratio)

colors = [
    "#5C2407",  # 更深棕橙（新增）
    "#80350E",
    "#C04F15",
    "#EC844A",
    "#F2AA84",
    "#F6C6AA",  # 更浅橙（新增）
    "#FAE3D6"   # 接近白的暖橙（新增）
]

# =========================
# 画 donut
# =========================
fig, ax = plt.subplots(figsize=(7,7))

explode = [0.03] * len(counts)

wedges, texts = ax.pie(
    counts,
    colors=colors,
    startangle=90,
    explode=explode,
    counterclock=False,
    wedgeprops=dict(width=0.1)
)

# =========================
# 添加向外发射的虚线
# =========================
r_inner = 0.8         # 起点
r_outer = 1.3      # 终点

for w in wedges:
    
    angle = w.theta1
    angle_rad = np.deg2rad(angle)

    x1 = r_inner * np.cos(angle_rad)
    y1 = r_inner * np.sin(angle_rad)

    x2 = r_outer * np.cos(angle_rad)
    y2 = r_outer * np.sin(angle_rad)

    ax.plot(
        [x1, x2],
        [y1, y2],
        linestyle="--",
        linewidth=1.5,
        color="black",
        alpha=0.6
    )

ax.set(aspect="equal")

plt.tight_layout()
plt.savefig("../plots/fig3_c.svg", dpi=600, bbox_inches="tight")
plt.show()
```
